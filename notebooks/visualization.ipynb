```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('../results/arxiv/edge/trends.csv')
```

```python
df.head()
```

```python
# Create a grouped bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
```

```python
grouped_data = df.groupby(['setting', 'no of hops', 'sampled nodes']).agg({'mean accuracy': 'mean', 'mean failure fraction': 'mean'}).reset_index()

# Set the style
sns.set(style="whitegrid")

# Create the grouped bar chart
plt.figure(figsize=(12, 6))
g = sns.catplot(
    x="setting",
    y="mean accuracy",
    hue="sampled nodes",
    col="no of hops",
    data=grouped_data,
    kind="bar",
    height=6,
)

# Add labels and title
g.set_axis_labels("Setting", "Mean Accuracy")
#g.set_titles("Mean Accuracy for Hop {col_name}, Sampled Nodes {hue_name}")
plt.subplots_adjust(top=0.85)
plt.suptitle("Mean Accuracy vs. Setting for Different Combinations")

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Set the style
sns.set(style="whitegrid")

# Create the grouped bar chart
plt.figure(figsize=(12, 6))
g = sns.catplot(
    x="setting",
    y="mean failure fraction",
    hue="sampled nodes",
    col="no of hops",
    data=grouped_data,
    kind="bar",
    height=6,
)

# Add labels and title
g.set_axis_labels("Setting", "Mean Failure Fraction")
#g.set_titles("Mean Accuracy for Hop {col_name}, Sampled Nodes {hue_name}")
plt.subplots_adjust(top=0.85)
plt.suptitle("Mean Failure Fraction vs. Setting for Different Combinations")

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig('mean_fail_edge.png')
```
